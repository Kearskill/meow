## Data preprocessing (state)

- replacing State code to State name

```python
df_state = df_dominant.merge(df_state, left_on='State', right_on='StateID', how='left') # Bring the StateName from df_
df_state
```

```python
df_state.nunique()
```

```python
df_state = df_state.drop(columns=['StateID'])
```

```python
df_state = df_state.rename(columns={'StateName':'State'})
```

```python
df_state
```

```python
import os
os.makedirs('output',exist_ok=True)
df_state.to_csv('output/test.csv',index=False)
print("Data Frame has been exported to 'output/test.csv'")
```

## Data Preprocessing (Breed)

```python
df_breed 
```

- to preprocess breed, we need to get breedID and type to match breed name

```python
df_merged_breed = df_dominant.merge(df_breed, left_on=['Type','Breed1'], right_on=['Type','BreedID'], how='left')
```

```python
df_merged_breed
```

```python
df_merged_breed = df_merged_breed.drop(columns='Type')
df_merged_breed
```

```python
null_count = df_merged_breed['BreedName'].isnull().sum()
print(null_count)
```

since only 17 out of 14992 (0.11%) is null, i will just drop it

```python
df_merged_breed = df_merged_breed.dropna(subset='BreedName')
null_count = df_merged_breed['BreedName'].isnull().sum()
print(null_count)
```

```python
import os
os.makedirs('output', exist_ok=True)
df_merged_breed.to_csv('output/breed.csv', index=False)
```

# changing gender. 1 male, 2 female, 3 mixed 


```python
df_gender = df_merged_breed
df_gender['Gender'] = df_gender['Gender'].replace({1:'Male',2:'Female',3:'Mixed'})
df_gender
```

```python
# checking for issues
df_gender.nunique()
```

gender settled, no worries

```python
import os
os.makedirs('output',exist_ok=True)
df_gender.to_csv('output/gender.csv', index=True)
```

accidentally dropped the Type, merging back from dominant(master df)

```python
df_color = df_gender
color_mapping = {1:'Black',2:'Brown',3:'Golden',4:'Yellow',5:'Cream',6:'Gray',7:'White'}

df_color[['Color1','Color2','Color3']] = df_color[['Color1','Color2','Color3']].replace(color_mapping)
df_color
```

```python
# check for issue
zero_count = (df_color['Color1'] == 0).sum()
print(zero_count)
zero_count = (df_color['Color2'] == 0).sum()
print(zero_count)
zero_count = (df_color['Color3'] == 0).sum()
print(zero_count)
```

```python
# apply color1 to color2
df_color['Color2'] = df_color.apply(lambda row: row['Color1'] if row['Color2'] == 0 else row['Color2'], axis=1)
df_color['Color3'] = df_color.apply(lambda row: row['Color2'] if row['Color3'] == 0 else row['Color3'], axis=1)
df_color
```

```python
import os
os.makedirs('output',exist_ok=True)
df_color.to_csv('output/color.csv',index=True)
```

MaturitySize - Size at maturity (1 = Small, 2 = Medium, 3 = Large, 4 = Extra Large, 0 = Not Specified)


```python
df_maturity = df_color
```

```python
df_maturity['MaturitySize'] = df_maturity['MaturitySize'].replace({0:'Not Specified',1:'Small',2:'Medium',3:'Large',4:'Extra Large'})
df_maturity
```

```python
import os
os.makedirs('output',exist_ok=True)
df_maturity.to_csv('output/maturity.csv',index=True)
```
