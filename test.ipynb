## Data preprocessing

- replacing State code to State name

```python
df_state = df_dominant.merge(df_state, left_on='State', right_on='StateID', how='left') # Bring the StateName from df_
df_state
```

```python
df_state.nunique()
```

```python
df_state = df_state.drop(columns=['StateID'])
```

```python
df_state = df_state.rename(columns={'StateName':'State'})
```
